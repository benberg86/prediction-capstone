```python
%pylab inline
import matplotlib.pyplot as plt
import pandas as pd

import numpy as np
```

```python
#load BLS Jolt data files and join using pandas
#allitems.csv
allitems_df = pd.read_csv('data/allitems.csv')
allitems_df
```

```python
#series.csv
series_df = pd.read_csv('data/series.csv')
series_df
```

```python
#join all items and series data

joined_all_df = pd.merge(left=allitems_df,right=series_df, left_on='series_id', right_on='series_id')
print joined_all_df.shape
joined_all_df
```

```python
#dataelement.csv
dataelement_df = pd.read_csv('data/dataelement.csv')
dataelement_df
```

```python
joined_all_df = pd.merge(left=joined_all_df,right=dataelement_df, left_on='dataelement_code', right_on='dataelement_code')
print joined_all_df.shape
joined_all_df
```

```python
#region
region_df = pd.read_csv('data/region.csv')
#ratelevel
ratelevel_df = pd.read_csv('data/ratelevel.csv')
#industry
industry_df = pd.read_csv('data/industry.csv')
#print industry_df
#seasonal
seasonal_df = pd.read_csv('data/seasonal.csv')
#print seasonal_df
#footnote
footnote_df = pd.read_csv('data/footnote.csv')
```

```python
#join region
joined_all_df = pd.merge(left=joined_all_df,right=region_df, left_on='region_code', right_on='region_code')
#join ratelevel
joined_all_df = pd.merge(left=joined_all_df,right=ratelevel_df, left_on='ratelevel_code', right_on='ratelevel_code')
#join industry
joined_all_df = pd.merge(left=joined_all_df,right=industry_df, left_on='industry_code', right_on='industry_code')
#join seasonal
joined_all_df = pd.merge(left=joined_all_df,right=seasonal_df, left_on='seasonal', right_on='industry_code')
print joined_all_df.shape
joined_all_df
```

```python
#focus on possible columns of interest
joined_all_df.drop(['footnote_codes_x','footnote_codes_y','industry_code_x','region_code','dataelement_code','ratelevel_code',
                    'footnote_code_y','display_level_x','selectable_x','sort_sequence_x','display_level_y',
                    'selectable_y','sort_sequence_y','display_level_x','selectable_x','industry_code_y'], inplace=True,axis=1,errors='ignore'
                    )
```

```python
#write to csv for now
joined_all_df.to_csv('data/joined_data_all.csv')
```
